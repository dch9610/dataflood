```python
import pandas as pd
```

```python
data = pd.read_excel('xycode.xlsx')
```

```python
data.info()
```

```python
data['CODE'] = data['행정구역코드'].apply(str)
```

```python
busan_data = data[data['CODE'].str.startswith('26')]
```

```python
busan_data = busan_data[['2단계','3단계','격자 X', '격자 Y', 'CODE']]
```

```python
busan_data = busan_data.fillna(0)
busan_data = busan_data[busan_data['3단계']==0]
```

```python
busan_data = busan_data.drop(columns='3단계')
busan_data = busan_data.drop(index=451)
busan_data
```

```python
busan_data = busan_data.rename(columns={'2단계':'ZONE','격자 X':'X','격자 Y' : 'Y'})
```

```python
busan = busan_data.reset_index()
busan = busan.drop(columns='index')
```

```python
base_dong = pd.read_csv('./base_dong.csv')
```

```python
base_geo_data = pd.merge(base_dong,busan,on='ZONE', how='left')
```

```python
base_geo_data
```

```python
base_geo_data = base_geo_data.drop(columns=['Unnamed: 0','CODE'])
```

```python
base_geo_data.shape
```

```python
base_geo_data.to_csv('base_data.csv', encoding='utf-8')
```
